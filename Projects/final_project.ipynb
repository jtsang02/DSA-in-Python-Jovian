1. Initialize an array of size 4 with all 1s representing the occurence of each letter in 'zero'.
2. Count how many times each letter needs to be repeated.
3. Initialize a current variable to 1 to count the number of subsequences of 'zero'.  `current` should be less than n.
4. Track the position of the char we are iterating thru in the array.  Reset this pointer to 0 if it reaches the end of the array.  


5. Return the sum of the array.

###  4. Implement the solution and test it using example inputs. Fix bugs, if any.

```python
def len_shortest_str(n):
    
    # initialize list of size 4 of all 1s
    arr = [1] * 4
    
    # pointer to loop thru each letter in 'zero' and
    # counter to track the current subsequences
    idx, cur_ss = 0, 1
    
    # iterate thru list repeatedly while count of subsequences is less than n
    while (cur_ss < n):
        
        # find the remaining count of subsequences to have
        cur_ss = cur_ss / arr[idx]
        
        # increment count of each char
        arr[idx] += 1
        
        # update number of subsequences i.e if cur_ss is 1 and 'zerooo' => 3 x 1 
        cur_ss = cur_ss * arr[idx]
        
        # increment or reset position
        idx = idx + 1 if idx < len(arr) else 0
    
    # return all the occurances of each letter
    print(arr)
    return sum(arr)
```
